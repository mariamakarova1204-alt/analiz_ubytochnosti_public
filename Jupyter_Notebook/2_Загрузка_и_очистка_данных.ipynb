```python
#делаем чистую выручку без НДС
df90_2025['Выручка_без_НДС'] = df90_2025['Сумма_кредит'] / 1.2
```

```python
#делаем abc-анализ
df_abc_2025 = (
    df90_2025
    .groupby('Контрагент')['Выручка_без_НДС']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
#считаем долю и накопленную долю
df_abc_2025['Доля_%'] = (
    df_abc_2025['Выручка_без_НДС']
    / df_abc_2025['Выручка_без_НДС'].sum()
    * 100
)
df_abc_2025['Накопленная_доля_%'] = df_abc_2025['Доля_%'].cumsum()
```

```python
#присваиваем группы
df_abc_2025['ABC'] = 'C'
df_abc_2025.loc[
    df_abc_2025['Накопленная_доля_%'] <= 80,
    'ABC'
] = 'A'
df_abc_2025.loc[
    (df_abc_2025['Накопленная_доля_%'] > 80)
    & (df_abc_2025['Накопленная_доля_%'] <= 95),
    'ABC'
] = 'B'
```

```python
#проверяем
df_abc_2025.groupby('ABC').agg({
    'Контрагент': 'count',
    'Выручка_без_НДС': 'sum',
    'Доля_%': 'sum'
})
```

```python
#создадим месяц
df90['Дата'] = pd.to_datetime(
    df90['Дата'],
    dayfirst=True,
    errors='coerce'
)
df90['Месяц'] = df90['Дата'].dt.month
df90[['Дата', 'Год', 'Месяц']].head()
```

```python
#посчитаем выручку в каждом месяце
df_revenue_accounts = ['90.01', '90.1.1']
df_monthly_revenue = (
    df90[
        df90['Счет_кредит'].isin(['90.01', '90.1.1'])
    ]
    .groupby(['Год', 'Месяц'])['Сумма_кредит']
    .sum()
    .reset_index(name='Выручка')
)
df_monthly_revenue.head()
```

```python
#посчитаем затраты в каждом месяце
df_monthly_expenses = (
    df_expenses_new
    .groupby(['Год', 'Месяц'])['Сумма_дебет']
    .sum()
    .reset_index(name='Затраты')
)
df_monthly_expenses.head()
```

```python
#сделаем такой же формат периода, как в затратах
df90['Дата'] = pd.to_datetime(df90['Дата'])
df_monthly_revenue = (
    df90[
        df90['Счет_кредит'].isin(['90.01', '90.1.1'])
    ]
    .assign(
        Месяц=lambda x: x['Дата'].dt.to_period('M')
    )
    .groupby(['Год', 'Месяц'])['Сумма_кредит']
    .sum()
    .reset_index(name='Выручка')
)
df_monthly_revenue.head()
```

```python
#посчитаем выручку без ндс
df_monthly_revenue['Выручка_без_НДС'] = (df_monthly_revenue['Выручка'] / 1.2)
```

```python
#теперь сделаем общим dataframe по затратам и выручке
powerbi_analiz = df_monthly_revenue.merge(
    df_monthly_expenses,
    on=['Год', 'Месяц'],
    how='outer'
)
powerbi_analiz['Прибыль'] = (
    powerbi_analiz['Выручка_без_НДС']
    - powerbi_analiz['Затраты']
)
powerbi_analiz['Рентабельность_%'] = (
    powerbi_analiz['Прибыль']
    / powerbi_analiz['Выручка_без_НДС']
    * 100
)
powerbi_analiz = powerbi_analiz.sort_values(
    ['Год', 'Месяц']
)
powerbi_analiz
```
